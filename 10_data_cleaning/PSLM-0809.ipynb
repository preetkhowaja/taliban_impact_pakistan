## PSLM 2008-2009

```python
import pandas as pd

C = pd.read_stata(
    "/Users/johnowusuduah/Downloads/data in stata 2008-09/0809_sec_c.dta"
)
print(C.shape)
C.sample(5)
```

```python
# Reduce to the questions of interest (4 questions):

C["ever_admitted"] = C["scq03"]
C["max_level_acheived"] = C["scq04"]
C["currently_enrrolled"] = C["scq05"]
C["why_not"] = C["scq09a"]

C_relevant = C[
    [
        "hhcode",
        "idc",
        "ever_admitted",
        "max_level_acheived",
        "currently_enrrolled",
        "why_not",
        "region",
        "province",
    ]
].copy()
```

```python
C_relevant
```

```python
B = pd.read_stata(
    "/Users/johnowusuduah/Downloads/data in stata 2008-09/0809_sec_b.dta"
)
B.sample(5)
```

```python
# Change the names of the questions for readability
B['sex'] = B['sbq01']
B['marital_status'] = B['sbq05']

B_relevant = B[['hhcode', 'idc', 'age', 'sex', 'marital_status']].copy()
```

```python
# Made the Household Code column a str type
B["hhcode"] = B["hhcode"].astype("str")
C["hhcode"] = C["hhcode"].astype("str")
```

```python
# MERGE
years08_09 = pd.merge(B_relevant, C_relevant, on=["hhcode", "idc"], how='inner', indicator=True)

# Evaluate merge 
years08_09._merge.value_counts()
```

```python
# Select people in school enrollment age (I would say 4-10 years old)
years08_09 = years08_09.loc[(years08_09["age"] >= 4)&(years08_09["age"] <= 10)]
```

```python
years08_09.head()
```

```python
years08_09.province.value_counts()
```

```python
# Lets see which hh codes we have:
years08_09["hhcode"] = years08_09["hhcode"].astype("str")
years08_09['subprovince code']= years08_09['hhcode'].apply(lambda x: x[0:4])
# remove the decimal characters from the string values
years08_09['hhcode']= years08_09['hhcode'].apply(lambda x: x[0:-2])
years08_09["subprovince code"] = years08_09["subprovince code"].astype("int")
# reformat marital status values
years08_09["marital_status"] = years08_09["marital_status"].astype("str")
years08_09.loc[(years08_09["marital_status"] == "unmarried"), 'marital_status'] = 1
years08_09.loc[(years08_09["marital_status"] == "married"), 'marital_status'] = 2
years08_09.loc[(years08_09["marital_status"] == "widow/widower"), 'marital_status'] = 3
years08_09.loc[(years08_09["marital_status"] == "divorced"), 'marital_status'] = 4
years08_09.loc[(years08_09["marital_status"] == "only nikha has been solemanised,rukhsati not taken"), 'marital_status'] = 5
years08_09.head()
```

```python
# Now we import the stratum name - code data
strat_name = pd.read_csv(
    "/Users/johnowusuduah/Downloads/data in stata 2008-09/stratum0809.csv"
)
strat_name.sample(5)
```

```python
# MERGE
years08_09_ = pd.merge(years08_09, strat_name, on=["subprovince code"], how='inner')

# Evaluate merge 
years08_09_._merge.value_counts()
```

```python
years08_09_.sample(5)
```

```python
# FINAL DATA PRE-PROCESSING
# subset columns of interest
years08_09_f = years08_09_[["hhcode","age","idc","sex","marital_status","ever_admitted","currently_enrrolled","region","subprovince code","province_y","subprovince"]].copy()
# rename index and province variables
years08_09_f = years08_09_f.rename(columns={"province_y":"province"})
# add year to data
years08_09_f["year"] = 2008
# change NWFP to KPK
years08_09_f.loc[years08_09_f["province"] == "NWFP", "province"] = "KPK"
# reformat columns/variables for consistency across all data sets
years08_09_f["sex"] = years08_09_f["sex"].astype("category")
years08_09_f["marital_status"] = years08_09_f["marital_status"].astype("int8")
years08_09_f["ever_admitted"] = years08_09_f["ever_admitted"].astype("str")
years08_09_f["currently_enrrolled"] = years08_09_f["currently_enrrolled"].astype("str")
years08_09_f["region"] = years08_09_f["region"].astype("str")
# make sure Islamabad is Punjab
years08_09_f.loc[(years08_09_f["province"] == "Islamabad"), 'province'] = "Punjab "
# preview sample of data set
years08_09_f.sample(5)
```
