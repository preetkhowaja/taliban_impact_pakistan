## Ingest Education Data Set

```python
import pandas as pd
pd.set_option('display.max_columns', None)

C = pd.read_stata(
    "/Users/johnowusuduah/Downloads/data_in_stata_2011-2012/sec_2a.dta", convert_categoricals=False
    )
print(C.shape)
C.sample(5)
```

```python
# s2bq01 is the variable that has 
# 1 - never attended school/institution
# 2 - attended school/institution in the past
# 3 - currently attending school/institution
C.s2bq01.value_counts()
```

```python
# Required Columns from Education Section
# create "ever_admitted" column
yes_admitted = [2,3]
C.loc[C["s2bq01"].isin(yes_admitted), "ever_admitted"] = 'yes'
C.loc[C["s2bq01"] == 1, "ever_admitted"] = 'no'
# create "currently_enrolled" column
C.loc[C["s2bq01"] == 3, "currently_enrolled"] = 'yes'
C.loc[C["s2bq01"] != 3, "currently_enrolled"] = 'no'

# Optional Columns from Education Section
# Note that "max_level_achieved" and "why_not" have integer encoded, refer to questionare
# create "max_level_achieved" column --> integer encoding already done
C["max_level_achieved"] = C["s2bq05"]

# create "why_not" column --> integer encoding already done
C["why_not"] = C["s2bq02"]

C.sample(5)
```

```python
# Reduce to the questions of interest (4 questions):

#C["ever_admitted"] = C["scq03"]
#C["max_level_acheived"] = C["scq04"]
#C["currently_enrrolled"] = C["scq05"]
#C["why_not"] = C["scq91"]

C_relevant = C[
    [
        "hhcode",
        "idc",
        "ever_admitted",
        "max_level_achieved",
        "currently_enrolled",
        "why_not",
        "region",
        "province",
    ]
].copy()
```

## Ingest Demography Data Set

```python
B = pd.read_stata(
    "/Users/johnowusuduah/Downloads/data_in_stata_2011-2012/roster.dta", convert_categoricals=False
)
B.sample(5)
```

```python
# Change the names of the questions for readability
B['sex'] = B['s1aq03']
#B['age'] = B['sbq04']
B['marital_status'] = B['s1aq06']

B_relevant = B[['hhcode', 'idc', 'age', 'sex', 'marital_status']].copy()
```

```python
# Made the Household Code column a str type
B["hhcode"] = B["hhcode"].astype("str")
C["hhcode"] = C["hhcode"].astype("str")
```

```python
# MERGE
years11_12 = pd.merge(B_relevant, C_relevant, on=["hhcode","idc"], how='inner', indicator=True)

# Evaluate merge 
years11_12._merge.value_counts()
```

```python
# Select people in school enrollment age (I would say 4-15 years old)
years11_12 = years11_12.loc[(years11_12["age"] >= 4)&(years11_12["age"] <= 15)]
```

```python
years11_12.head()
```

```python
years11_12.province.value_counts()
```

```python
# Lets see which hh codes we have:
years11_12["hhcode"] = years11_12["hhcode"].astype("str")
years11_12['subprovince code']= years11_12['hhcode'].apply(lambda x: x[0:4])
years11_12["subprovince code"] = years11_12["subprovince code"].astype("int")
# reformat region values
years11_12.loc[(years11_12["region"] == 1), 'region'] = 'urban'
years11_12.loc[(years11_12["region"] == 2), 'region'] = 'rural'
# reformat sex values
years11_12.loc[(years11_12["sex"] == 1), 'sex'] = 'male'
years11_12.loc[(years11_12["sex"] == 2), 'sex'] = 'female'
# reformat marital status values
years11_12.loc[(years11_12["marital_status"] == 1), 'marital_status'] = 'unmarried/never married'
years11_12.loc[(years11_12["marital_status"] == 2), 'marital_status'] = 'married'
years11_12.loc[(years11_12["marital_status"] == 3), 'marital_status'] = 'widow/widower'
years11_12.loc[(years11_12["marital_status"] == 4), 'marital_status'] = 'divorced'
years11_12.loc[(years11_12["marital_status"] == 5), 'marital_status'] = 'Nikkah solemnised Rukshati not yet'

years11_12.head()
```

```python
# Now we import the stratum name - code data
strat_name = pd.read_csv(
    "/Users/johnowusuduah/Downloads/data_in_stata_2011-2012/stratum1112.csv"
).iloc[:,0:3]

strat_name.sample(5)
```

```python
# MERGE
years11_12_ = pd.merge(years11_12, strat_name, on=["subprovince code"], how='inner')

# Evaluate merge 
years11_12_._merge.value_counts()
```
