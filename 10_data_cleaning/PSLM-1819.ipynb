## Cleaning File for 2018-19 Data


```python
import pandas as pd
import numpy as np
```

```python
roster_1819 = pd.read_stata('/Users/preetkhowaja/Documents/midssp2022/unifying/uds-2022-ids-701-team-3/00_source_data/1819_sec_1a.dta')
roster_1819.head()
```

```python
# renaming relevant variables from roster
roster_1819['sex'] = roster_1819['s1aq04']
roster_1819['marital_status'] = roster_1819['s1aq07']
roster_1819.sample(4)
```

```python
roster_1819.province.value_counts()
```

```python
B_relevant = roster_1819[['hhcode', 'idc','age', 'sex', 'marital_status']].copy()
B_relevant.sample(5)
```

```python
data_1819 = pd.read_stata('/Users/preetkhowaja/Documents/midssp2022/unifying/uds-2022-ids-701-team-3/00_source_data/1819_sec_2ab.dta')
data_1819.head()
```

```python
data_1819.s2bq01.value_counts()
```

```python
# Create variable for ever_admitted and currently attending
data_1819['ever_admitted'] = np.where(data_1819.s2bq01 == 'never attended school', 'no', 'yes')
data_1819.ever_admitted.value_counts()
data_1819['currently_enrolled'] = np.where(data_1819.s2bq01 == 'currently attending school', 'yes', 'no')
data_1819.currently_enrolled.value_counts()
```

```python
C_relevant = data_1819[
    [
        "hhcode",
        "idc",
        "ever_admitted",
        "currently_enrolled",
        "region",
        "province"
    ]
].copy()

C_relevant.sample(5)
```

```python
# MERGE 
years18_19 = pd.merge(B_relevant, C_relevant, on=["hhcode", 'idc'], how='inner', indicator=False)
# Evaluate merge by checking value counts for currently enrolled individuals
years18_19.currently_enrolled.value_counts()
```

```python
# Select people in school enrollment age (I would say 4-10 years old)
years18_19 = years18_19.loc[(years18_19["age"] >= 4)&(years18_19["age"] <= 15)]
```

```python
years18_19.province.value_counts()
```

```python
# Re-spell the names
years18_19['province'].replace('punjab', 'Punjab', inplace = True)
years18_19['province'].replace('khyber pakhtunkhwa', 'KPK', inplace = True)
years18_19['province'].replace('sindh', 'Sindh', inplace = True)
years18_19['province'].replace('balochistan', 'Balochistan', inplace = True)
years18_19.province.value_counts()
```

```python
# Lets see which hh codes we have:
years18_19["hhcode"] = years18_19["hhcode"].astype("str")
years18_19['subprovince code']= years18_19['hhcode'].apply(lambda x: x[0:4])
years18_19["subprovince code"] = years18_19["subprovince code"].astype("int")
years18_19.head()
```

```python
stratum_1819 = pd.read_csv('/Users/preetkhowaja/Documents/midssp2022/unifying/uds-2022-ids-701-team-3/00_source_data/stratum_1819.csv')
stratum_1819 = stratum_1819[['subprovince', 'subprovince code']]
stratum_1819.sample(2)
```

```python
# Merge stratum data with edu and sex data
years18_19_ = pd.merge(years18_19, stratum_1819, on=["subprovince code"], how='inner', indicator=True)

# Evaluate merge 
years18_19_._merge.value_counts()
```

```python
years18_19_.drop(columns= ['_merge'], inplace = True)
years18_19_.head()
```

```python
# Add year indicator for 2018
years18_19_['Year'] = 2018
```

```python
years18_19_.marital_status.value_counts()
```

```python
# Recode marital status
years18_19_.marital_status.replace('unmarried / never married', 1, inplace = True)
years18_19_.marital_status.replace('currently married', 2, inplace = True)
years18_19_.marital_status.replace('widow / widower', 3, inplace = True)
years18_19_.marital_status.replace('separated', 4, inplace = True)
years18_19_.marital_status.replace('divorced', 4, inplace = True)
years18_19_.marital_status.replace('nikkah solemnized but rukhsati not taken place', 5, inplace = True)
years18_19_.marital_status.value_counts()
```

```python
# Save to csv
years18_19_.to_csv('/Users/preetkhowaja/Documents/midssp2022/unifying/uds-2022-ids-701-team-3/10_data_cleaning/clean_preet/years18_19_.csv', index_label='id')
```
